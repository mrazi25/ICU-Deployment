```python
import pandas as pd
from itertools import product

phys_act = [0, 2, 1, 1, 2, 1, 5, 5, 0, 9, 12]
ai_act = [0, 2, 1, 1, 1, 1, 21, 15, 2, 5, 2]
doses_intensity = ["zero", "low", "medium", "high", "very high"]
pair_of_act = list(product(doses_intensity, repeat=2))

df_data = {"Physician": [], "AI (Deep Q Network)": [], "Intrafluid": [], "Vasopressor": []}
df_data_2 = {"Four-Hourly Timestamp": []}

# Your loop to fill the data
for i in range(len(phys_act)):
    df_data_2["Four-Hourly Timestamp"].append(i + 1)
    df_data["Physician"].append(pair_of_act[phys_act[i]])
    df_data["AI (Deep Q Network)"].append(pair_of_act[ai_act[i]])
    df_data["Intrafluid"].append(f"Intrafluid {i + 1}")
    df_data["Vasopressor"].append(f"Vasopressor {i + 1}")

# Create MultiIndex for both rows and columns
rows_multiindex = pd.MultiIndex.from_product([doses_intensity, range(1, len(phys_act) + 1)],
                                             names=["Dose Intensity", "Four-Hourly Timestamp"])
columns_multiindex = pd.MultiIndex.from_product([["Physician", "AI (Deep Q Network)"],
                                                ["Intrafluid", "Vasopressor"]],
                                               names=["", ""])

# Create DataFrames with the MultiIndex
df = pd.DataFrame(df_data, index=rows_multiindex, columns=columns_multiindex)
df_2 = pd.DataFrame(df_data_2["Four-Hourly Timestamp"], index=rows_multiindex, columns=[("Four-Hourly Timestamp", "")])

# Display the resulting DataFrame
df_result = pd.concat([df_2, df], axis=1)
print(df_result)
```

```python
phys_act = [0, 2, 1, 1, 2, 1, 5, 5, 0, 9, 12]
ai_act = [0, 2, 1, 1, 2, 1, 5, 5, 0, 9, 12]
doses_intensity = ["zero", "low", "medium", "high", "very high"]
pair_of_act = list(product(doses_intensity, repeat=2))

# Create a MultiIndex for columns
columns = pd.MultiIndex.from_product([["Physician", "AI (Deep Q Network)"], doses_intensity], names=["Activity", "Intensity"])

# Create an empty DataFrame with the MultiIndex columns
df = pd.DataFrame(index=range(len(phys_act)), columns=columns)

# Fill the DataFrame with data
for i in range(len(phys_act)):
    for intensity in doses_intensity:
        df.loc[i, ("Physician", intensity)] = phys_act[i] if intensity == pair_of_act[i][0] else None
        df.loc[i, ("AI (Deep Q Network)", intensity)] = ai_act[i] if intensity == pair_of_act[i][0] else None

# Display the DataFrame using st.dataframe
df
```
