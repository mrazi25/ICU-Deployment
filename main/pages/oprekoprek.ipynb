```python
save_indexes = []
i_index = -1+1
for i, row in data_test.iterrows():
    if data_test.loc[i, 'icustayid'] == 269.0:
        save_indexes.append(i)
```

```python
phys_act = []
ai_act = []

for i in range(len(save_indexes)):
    phys_act.append(phys_action_test[save_indexes[i]])
    print(f"Physician action is giving category of {phys_act[i]} for combination of vasopressor and intrafluid doses")
    ai_act.append(agent_action_test[save_indexes[i]])
    print(f"AI action is giving category of {ai_act[i]} for combination of vasopressor and intrafluid doses")
    print("===========================================================================================================\n")
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Data
x = range(len(ai_act))  # Anggap saja data1 dan data2 memiliki panjang yang sama
y1 = ai_act
y2 = phys_act

# Plotting garis
plt.plot(x, y1, label='AI', marker='o')
plt.plot(x, y2, label='Physician', marker='o')

# Menambahkan rentang nilai y hingga 24
plt.ylim(0, 24)

# Menambahkan label dan judul
plt.xlabel('State')
plt.ylabel('Action Choosen')
plt.title('Line Chart with Two Lines and Dots')

# Menambahkan legenda
plt.legend()

# Menambahkan grid
plt.grid(True)

# Menampilkan plot
plt.show()
```
